### Importando as bibliotecas

```python
import numpy as np
import pandas as pd
from abc_analysis import abc_analysis
```

```python
### ler o dado
df=pd.read_csv("vendas.csv",sep=";")
```

```python
df
```

```python
#ver o tipo de dado que tem
df.dtypes
```

```python
#fazer copia de segurança
df_edited=df.copy()
```

```python
#removendo os na
df_edited.dropna(inplace=True)
```

```python
df_edited
```

```python
df_edited.iloc[:,2:]
```

```python
#tratamento dos dados

#remover cifrões e tracinhos ou seja tudo que é numerico
table_number=df_edited.iloc[:,2:].replace(r'\D','',regex=True)
table_number=table_number.replace("",0)
table_number=table_number.astype(int)
```

```python
table_number
```

```python
### juntar as primeiras colunas da tabela de origem com a tabela numérica

table_complete=pd.concat([df_edited.iloc[:,:2],table_number],axis=1)
```

```python
table_complete
```

```python
#criando coluna nova
table_complete['receita_13_meses']=table_complete.iloc[:,5:(5+13)].sum(axis=1)
```

```python
table_complete
```

```python
#fazer a bruxaria

abc=abc_analysis(table_complete['receita_13_meses'])
```

```python
abc
```

```python
abc_analysis(table_complete['receita_13_meses'],True)
```

```python
### fazer classificacao dos dados

# indices

a_index=abc['Aind']
b_index=abc['Bind']
c_index=abc['Cind']

# New Column indicating A, B, or C
cond_list = [table_complete.index.isin(a_index),
             table_complete.index.isin(b_index),
             table_complete.index.isin(c_index)]

choice_list = ['A','B','C']

table_complete['abc'] = np.select(cond_list, choice_list)
table_complete.sort_values('receita_13_meses', ascending=False)
```
